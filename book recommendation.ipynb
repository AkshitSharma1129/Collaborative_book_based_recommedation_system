```python
ratings_with_name = ratings.merge(books,on='ISBN')
# merge books and ratings on isbn 
```

```python
ratings_with_name.head()
```

```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'total_ratings','Book-Title_x':'Book_Title'},inplace=True)
```

```python
num_rating.head()
```

```python
avg_rating = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'total_ratings','Book-Title_x':'Book_Title'},inplace=True)
avg_rating.rename(columns={'total_ratings':'avg_ratings'},inplace=True)
avg_rating.head()
```

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
popular_df.head()
```

```python
popular_df[popular_df['total_ratings']>=50].sort_values(by='total_ratings', ascending=[False]).head()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
# boolean indexing
num_users = x[x].index
# each element is True if the user has made more than 200 ratings and False otherwise.
# So, num_users will be a Series containing the 'User-ID's of users who have made more than 200 ratings in the ratings_with_name DataFrame.
```

```python
y = ratings_with_name['User-ID'].isin(num_users)
filtered_rating = ratings_with_name[y]
filtered_rating.head()
# The resulting filtered_rating DataFrame will contain only the rows where the 'User-ID' matches one of the 'User-ID's in num_users, effectively filtering out the ratings made by users with fewer than 200 ratings.
```

```python
z = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books =  z[z].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.head().iloc[:,[0,1,2,3,4]]
# woh users hai jinhone ne 200 se zayada books ko rate kiya and woh books hai jinpe 50 se zayada rating hui hai
```

```python
pt.fillna(0,inplace=True)
pt.head().iloc[:,[0,1,2,3,4]]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# calculate the cosine similarity between two or more vectors. In the context of machine learning and data analysis, cosine similarity is commonly used to measure the similarity between two data samples, represented as vectors
```

```python
# calculate the cosine similarity between two or more vectors. In the context of machine learning and data analysis, cosine similarity is commonly used to measure the similarity between two data samples, represented as vectors
```

```python
similarity_score = cosine_similarity(pt)
# humne 706 books ka euclidian distance nikala hai 706 books ke saath
```

```python
list(enumerate(similarity_score[0]))
# 1984 ka 0th index wale ke saath ka score
# 1984 ka 1st index wale ke saath ka score
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    # woh book kis index pe hai
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('Message in a Bottle')
```
